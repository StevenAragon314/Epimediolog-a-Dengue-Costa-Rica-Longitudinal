# Cálculo de la tasa de incidencia y el cambio porcentual y número de casos por cantón

```python
import pandas as pd
import numpy as np
```

```python
casos = pd.read_csv("../data/cases_dengue_rr.csv")
pob = pd.read_csv("../data/pob_int_rr.csv")
```

```python
casos.drop(["0", "1", "2", "3", "4", "5", "6", "7", "99", "98"], axis = 1, inplace= True)
```

```python
pob.drop(["Unnamed: 0", "  0", "612", "613"], axis = 1, inplace= True)
```

```python
pob = pob.rename({"week": "semana"}, axis = 1)
```

# Tasa de incidencia

  $$\text{Tasa de Incidencia} = \left( \frac{\text{Casos Nuevos de Dengue en la Semana } w}{\text{Población Total en el Cantón } c} \right) \times 100,000$$

```python
casos_idx = casos.set_index('semana')
pob_idx = pob.set_index('semana')
```

```python
incidencia = (casos_idx / pob_idx) * 100000
```

```python
incidencia.head()
```

```python
incidencia.to_csv("../data/incidencia.csv")
```

# Cambio porcentual y número de casos por cantón

$$\Delta\% = \left( \frac{\text{Casos en la Semana } w - \text{Casos en la Semana } w-1}{\text{Casos en la Semana } w-1} \right) \times 100$$

```python
cambio =  pd.DataFrame(np.where((casos_idx.shift(1) == 0) & (casos_idx == 0), 0, casos_idx.pct_change() * 100 ), index=casos_idx.index, columns=casos_idx.columns)
cambio = cambio.replace([np.inf, -np.inf], np.nan)
cambio = cambio.iloc[1:cambio.shape[0], 0:cambio.shape[1]]
```

```python
cambio.to_csv("../data/cambio.csv")
```
